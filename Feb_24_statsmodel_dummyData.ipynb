## Linear Regression using statsmodel for csv file--> 'dummy.csv'

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# reading data
data = pd.read_csv('dummy.csv')
```

```python
data  # data showing numerical (SAT,GPA) and categorical values(Attendance)
```

```python
# converting categoricalvalues into numerical value to interpret data
#using .map() function

data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
x = data[['SAT','Attendance']]
```

```python
x
```

```python
y = data['GPA']
```

```python
y
```

```python
# importing statsmodel
import statsmodels.api as sm
```

```python
# initialising constant--->default as 1.0
x = sm.add_constant(x)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model
```

```python
model.summary()
```
